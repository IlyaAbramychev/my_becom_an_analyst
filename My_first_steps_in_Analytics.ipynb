### –ó–∞–¥–∞–Ω–∏–µ 1.4

–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –ø–µ—Ä–µ–≤–æ—Ä–∞—á–∏–≤–∞—Ç—å —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞, —Ç–æ –µ—Å—Ç—å –∑–∞–ø–∏—Å—ã–≤–∞—Ç—å –∏—Ö —Ü–∏—Ñ—Ä—ã –≤ –æ–±—Ä–∞—Ç–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ. –ù–∞–ø—Ä–∏–º–µ—Ä, —á–∏—Å–ª–æ 734 –ø—Ä–µ–≤—Ä–∞—Ç–∏—Ç—Å—è –≤ 437. –ü–æ–∫–∞–∂–∏—Ç–µ —Ä–∞–±–æ—Ç—É —Ñ—É–Ω–∫—Ü–∏–∏ –Ω–∞ –ø—Ä–∏–º–µ—Ä–µ.



```python
#–ø—Ä–µ–∫–æ–¥
def reverse_int(x):
    
    negative_digit = x < 0
    if negative_digit:
        x = -x
        
    y = x % 10
    x2 = y
    x = x // 10
    
    while x > 0:
        y = x % 10
        x = x // 10
        x2 = x2 * 10
        x2 = x2 + y
        
    return x2

reverse_int(-123456789)
```
