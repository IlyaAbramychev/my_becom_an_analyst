
### Задание 1.4

Напишите функцию, которая будет переворачивать целые числа, то есть записывать их цифры в обратном порядке. Например, число 734 превратится в 437. Покажите работу функции на примере.



```python
#прекод
def reverse_int(x):
    
    negative_digit = x < 0
    if negative_digit:
        x = -x
        
    y = x % 10
    x2 = y
    x = x // 10
    
    while x > 0:
        y = x % 10
        x = x // 10
        x2 = x2 * 10
        x2 = x2 + y
        
    return x2

reverse_int(-123456789)
```
