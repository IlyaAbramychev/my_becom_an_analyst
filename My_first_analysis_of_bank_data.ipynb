```python
# –ú–µ—Å—Ç–æ –¥–ª—è –≤–∞—à–µ–≥–æ –∫–æ–¥–∞
import numpy as np

p0 = df['total_income'].quantile(0.0)
p14 = df['total_income'].quantile(0.14)
p34 = df['total_income'].quantile(0.34)
p78 = df['total_income'].quantile(0.78)
    
def income_level(row, p0, p14, p34, p78):
    
    """–§—É–Ω–∫—Ü–∏—è –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ –ª—é–¥–µ–π –ø–æ –∏—Ö –¥–æ—Ö–æ–¥—É:
       - –ë–µ–∑ –¥–æ—Ö–æ–¥–∞ ‚Äî –ª—é–¥–∏ –±–µ–∑ —Ä–∞–±–æ—Ç—ã –∏ —Å –Ω—É–ª–µ–≤—ã–º –¥–æ—Ö–æ–¥–æ–º.
       - –û—á–µ–Ω—å –Ω–∏–∑–∫–∏–π –¥–æ—Ö–æ–¥ ‚Äî –ª—é–¥–∏, –ø–æ–ª—É—á–∞—é—â–∏–µ –Ω–∏–∂–µ 14 –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—è –æ—Ç –æ–±—â–µ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–æ—Ö–æ–¥–∞.
       - –ù–∏–∑–∫–∏–π –¥–æ—Ö–æ–¥ ‚Äî –ª—é–¥–∏, –ø–æ–ª—É—á–∞—é—â–∏–µ –º–µ–∂–¥—É 14 –∏ 34 –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—è–º–∏ –æ—Ç –æ–±—â–µ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–æ—Ö–æ–¥–∞.
       - –°—Ä–µ–¥–Ω–∏–π –¥–æ—Ö–æ–¥ ‚Äî –ª—é–¥–∏, –ø–æ–ª—É—á–∞—é—â–∏–µ –º–µ–∂–¥—É 34 –∏ 78 –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—è–º–∏ –æ—Ç –æ–±—â–µ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–æ—Ö–æ–¥–∞.
       - –í—ã—Å–æ–∫–∏–π –¥–æ—Ö–æ–¥ ‚Äî –ª—é–¥–∏, –ø–æ–ª—É—á–∞—é—â–∏–µ –±–æ–ª—å—à–µ 78 –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—è –æ—Ç –æ–±—â–µ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–æ—Ö–æ–¥–∞."""
    
    total_income = row['total_income']
    
    if total_income <= p0:
        return '–ë–µ–∑ –¥–æ—Ö–æ–¥–∞'
    elif total_income <= p14:
        return '–û—á–µ–Ω—å –Ω–∏–∑–∫–∏–π –¥–æ—Ö–æ–¥'
    elif total_income <= p34:
        return '–ù–∏–∑–∫–∏–π –¥–æ—Ö–æ–¥'
    elif total_income <= p78:
        return '–°—Ä–µ–¥–Ω–∏–π –¥–æ—Ö–æ–¥'
    else:
        return '–í—ã—Å–æ–∫–∏–π –¥–æ—Ö–æ–¥'
    
df['calculation_income'] = df.apply(income_level, axis = 1, args = (p0, p14, p34, p78))
df.sample(5)
```


1. –†–∞–∑–¥–µ–ª–∏—Ç–µ –∫–ª–∏–µ–Ω—Ç–æ–≤ –Ω–∞ –¥–≤–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –ø–æ –≤–æ–∑—Ä–∞—Å—Ç—É: –¥–æ 40 –ª–µ—Ç –∏ –ø–æ—Å–ª–µ. –°–æ—Ö—Ä–∞–Ω–∏—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ –∫–æ–ª–æ–Ω–∫–µ `age_category`.



```python
# –ú–µ—Å—Ç–æ –¥–ª—è –≤–∞—à–µ–≥–æ –∫–æ–¥–∞
def customer_separation(row):
     
    """–§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è –ª—é–¥–µ–π –Ω–∞ –¥–≤–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏:
    - –î–æ 40 –ª–µ—Ç;
    - –ü–æ—Å–ª–µ 40 –ª–µ—Ç."""
    
    dob_years = row['dob_years']
    
    if dob_years < 40:
        return '–î–æ 40 –ª–µ—Ç'
    elif dob_years >= 40:
        return '–ü–æ—Å–ª–µ 40 –ª–µ—Ç'
    else:
        return '–ù–µ–≤–µ—Ä–Ω—ã–π –≤–æ–∑—Ä–∞—Å—Ç'
    
    
    
    
    
df['age_category'] = df.apply(customer_separation, axis = 1)
df.sample(5)
```


1. –†–∞–∑–¥–µ–ª–∏—Ç–µ –∫–ª–∏–µ–Ω—Ç–æ–≤ –Ω–∞ –Ω–µ—Å–∫–æ–ª—å–∫–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–π –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –¥–µ—Ç–µ–π: –±–µ–∑ –¥–µ—Ç–µ–π, –æ—Ç –æ–¥–Ω–æ–≥–æ –¥–æ –¥–≤—É—Ö, –æ—Ç —Ç—Ä—ë—Ö –∏ –±–æ–ª—å—à–µ. –°–æ—Ö—Ä–∞–Ω–∏—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ –∫–æ–ª–æ–Ω–∫–µ `childrens_category`.



```python
# –ú–µ—Å—Ç–æ –¥–ª—è –≤–∞—à–µ–≥–æ –∫–æ–¥–∞
def count_children(row):
    
    """–§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è –Ω–∞ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –¥–µ—Ç–µ–π:
    - –ë–µ–∑ –¥–µ—Ç–µ–π;
    - –û—Ç –æ—Ç–¥–Ω–æ–≥–æ –¥–æ –¥–≤—É—Ö –¥–µ—Ç–µ–π;
    - –û—Ç —Ç—Ä—ë—Ö –∏ –±–æ–ª—å—à–µ –¥–µ—Ç–µ–π."""
    
    children = row['children']
    
    if children <= 0:
        return '–ë–µ–∑ –¥–µ—Ç–µ–π'
    elif children >= 1 and children <= 2:
        return '–û—Ç –æ–¥–Ω–æ–≥–æ –¥–æ –¥–≤—É—Ö –¥–µ—Ç–µ–π'
    elif children >= 3:
        return '–û—Ç —Ç—Ä—ë—Ö –∏ –±–æ–ª—å—à–µ –¥–µ—Ç–µ–π'
    else:
        return '–ù–µ–≤–µ—Ä–Ω–æ'
    
    
df['childrens_category'] = df.apply(count_children, axis = 1)
df.sample(5)
```
